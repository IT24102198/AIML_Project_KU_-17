```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/KU-17_B01_Kandy_Uni/results/outputs/survey_no_outliers.csv')
```

```python
scaler = StandardScaler()

df['Age_standardized'] = scaler.fit_transform(df[['Age']])
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Age'], bins=30, edgecolor='black')
plt.title('Age Distribution (Before Scaling)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Age_standardized'], bins=30, edgecolor='black')
plt.title('Age Standardized Distribution')
plt.xlabel('Age (Standardized)')
plt.ylabel('Count')
plt.show()
```
